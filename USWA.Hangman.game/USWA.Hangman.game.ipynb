```python
class HangmanGame:
    def __init__(self):
        # Initialization for the HangmanGame class
        print("Welcome to Uswa's Hangman Game!")
        self.word_to_guess = input("Please enter the word to guess: ").lower()
        self.incorrect_guess = 6
        self.guesses = []
        self.attempts = 0

    def display_word(self):
        # Method to display the word with blanks for unguessed letters
        displayed = ""
        for letter in self.word_to_guess:
            if letter in self.guesses:
                displayed += letter
            else:
                displayed += "_"
            displayed += " "
        return displayed

    def play(self):
        # Method to play the Hangman game
        while self.attempts < self.incorrect_guess:
            guess = input("Enter your guess: ").lower()

            if len(guess) != 1 or not guess.isalpha():
                # Check if the input is a single letter
                print("Please enter a single letter.")
                continue

            if guess in self.guesses:
                # Check if the letter has already been guessed
                print("You already guessed that letter.")
            elif guess in self.word_to_guess:
                # If the guessed letter is in the word
                self.guesses.append(guess)
                print("You guessed correctly")
                word_display = self.display_word()
                print(f"Word to guess: {word_display}")
                if "_" not in word_display:
                    # Check if all letters have been guessed
                    print("Congratulations! You've guessed the word.")
                    self.play_again()
                    break
            else:
                # If the guessed letter is not in the word
                self.attempts += 1
                print(f"Incorrect guesses: {', '.join(self.guesses)}")

                if self.attempts == 1:
                    print(" + --- +\n | |\n O |\n |\n |\n |")
                elif self.attempts == 2:
                    print(" + --- +\n | |\n O |\n/|\n |\n |")
                elif self.attempts == 3:
                    print(" + --- +\n | |\n O |\n/|\\\n |\n |")
                elif self.attempts == 4:
                    print(" + --- +\n | |\n O |\n/|\\\n / \\\n |")

        if "_" in self.display_word():
            # If there are still blanks in the word
            print(f"Out of guesses! You lose. The word is \"{self.word_to_guess}\".")
            self.play_again()

    def play_again(self):
        # Method to ask if the player wants to play again
        play_again = input("Would you like to play again? (yes/no): ").lower()
        if play_again == "yes":
            # If the player wants to play again, restart the game
            self.__init__()
            self.play()
        else:
            # If the player doesn't want to play again, end the game
            print("Thanks for playing Uswa's Hangman Game!")


# Start the game
hangman_game = HangmanGame()
hangman_game.play()
```
