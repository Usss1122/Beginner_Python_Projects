```python
import random

class WordGuessGame:
    def __init__(self):
        # Initialize the game with a list of words, maximum attempts, and placeholders for the current word and display word
        self.words = ["python", "programming", "challenge", "hangman", "coding", "algorithm"]
        self.max_attempts = 5
        self.current_word = ""
        self.display_word = []
        self.guesses = []

    def select_word(self):
        # Select a random word for the player to guess
        self.current_word = random.choice(self.words)
        self.display_word = ["_"] * len(self.current_word)
        self.guesses = []

    def display_game_status(self):
        # Display the current status of the game
        print("\n--- Word Guess Game ---")
        print(" ".join(self.display_word))
        print("Incorrect Guesses:", " ".join(self.guesses))
        print(f"Attempts left: {self.max_attempts}")

    def make_guess(self):
        # Get a letter guess from the player
        while True:
            guess = input("Enter a letter guess: ").lower()
            if len(guess) == 1 and guess.isalpha():
                return guess
            else:
                print("Invalid input. Please enter a single letter.")

    def update_display_word(self, guess):
        # Update the display word with the correct guesses
        for i, letter in enumerate(self.current_word):
            if letter == guess:
                self.display_word[i] = guess

    def play_game(self):
        print("Welcome to the Word Guess Game!")

        self.select_word()

        while "_" in self.display_word and self.max_attempts > 0:
            self.display_game_status()

            guess = self.make_guess()

            if guess in self.guesses:
                print("You already guessed that letter. Try again.")
                continue

            self.guesses.append(guess)

            if guess in self.current_word:
                print("Correct guess!")
                self.update_display_word(guess)
            else:
                print("Incorrect guess!")
                self.max_attempts -= 1

        self.display_game_status()

        if "_" not in self.display_word:
            print("Congratulations! You guessed the word.")
        else:
            print(f"Sorry, you ran out of attempts. The word was '{self.current_word}'.")

if __name__ == "__main__":
    word_guess_game = WordGuessGame()
    word_guess_game.play_game()
```
